```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import json

# File to Load (Remember to change these)
tweets_data_to_load = "data/tweets_01-08-2021.csv"
```

```python
# Read the City Data
trump_tweets1 = pd.read_csv(tweets_data_to_load)
trump_tweets1.head()
```

```python
trump_tweets1.shape
```

```python
trump_tweets12 = trump_tweets1.loc[:, ["text", "date", "retweets", "favorites"]]
trump_tweets12
```

```python
result2= !garc userposts realdonaldtrump --gabs_after=2021-01-08
```

```python
list=[]
for gab in result2:
    list.append(json.loads(gab))
trump_tweets2 = pd.DataFrame(list)
trump_tweets2.head()
```

```python
for col in trump_tweets2.columns:
    print(f'"{col}"')
```

```python
trump_tweets2['created_at'] = trump_tweets2['created_at'].str.replace('T', ' ').str.replace('Z', '')
trump_tweets2['created_at'] = trump_tweets2['created_at'] .astype(str)
trump_tweets2['created_at'] = trump_tweets2['created_at'] .str.replace('\..*','')
trump_tweets22 = trump_tweets2.loc[:, ["body", "created_at", "reblogs_count", "favourites_count"]]
trump_tweets22.rename(columns={"body": "text", "created_at": "date", "reblogs_count": "retweets", "favourites_count": "favorites"}, inplace=True)
trump_tweets22
```

```python
trump_tweets22.dtypes
```

```python
frames = [trump_tweets12, trump_tweets22]
trump_tweets = pd.concat(frames)
trump_tweets['date'] = pd.to_datetime(trump_tweets.date)
trump_tweets.sort_values('date', inplace=True)
trump_tweets
```
